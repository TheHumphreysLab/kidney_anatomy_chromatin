```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_excel('cutrun_K-MetStat_Panel2.xlsx', index_col=0)
df
```

```python
df.index=['Cortex_H3K4me3_rep1','Cortex_H3K4me3_rep2','Cortex_IgG_rep1','Cortex_IgG_rep2',
              'Medulla_H3K4me3_rep1','Medulla_H3K4me3_rep2','Medulla_IgG_rep1','Medulla_IgG_rep2',
              'Papilla_H3K4me3_rep1','Papilla_H3K4me3_rep2','Papilla_IgG_rep1','Papilla_IgG_rep2']
df
```

```python
from  matplotlib.colors import LinearSegmentedColormap
```

```python
cmap_use=LinearSegmentedColormap.from_list('name',["sandybrown", "w", "blue"], N=256) 

plt.figure(dpi=300, figsize=(12, 6))
sns.set(font_scale=1.5)
heatmap = sns.heatmap(df, cmap=cmap_use, linewidths=1, xticklabels=True,
                      cbar_kws={'label': 'Histone modifications%', "shrink": 0.8},
                      annot=True, vmax=88, vmin=0, fmt=".1f",
                      annot_kws={"size": 12})

# Set the font size for x label
heatmap.set_xlabel(heatmap.get_xlabel(), fontsize=heatmap.get_xticklabels()[0].get_fontsize())

# Remove y-axis label
heatmap.set_ylabel('')

plt.xticks(rotation=60, fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.show()
```
