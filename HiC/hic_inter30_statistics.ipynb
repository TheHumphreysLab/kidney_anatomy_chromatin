```python
total=[]
alignable=[]
MAPQ=[]
hic=[]
inter=[]
intra=[]
short=[]
long=[]

#all files are available at this GitHub
for i in ['C1_inter_30.txt','C2_inter_30.txt','M1_inter_30.txt','M2_inter_30.txt','P1_inter_30.txt','P2_inter_30.txt']:
    file=open(i,'r')
    for line in file:
        line=line.strip().split(':')
        if 'Sequenced Read Pairs' in line[0]:
            total.append(int(line[1].strip().replace(",", "").split('(')[0].strip()))
        elif 'Alignable (Normal+Chimeric Paired)' in line[0]:
            alignable.append(int(line[1].strip().replace(",", "").split('(')[0].strip()))
        elif 'Below MAPQ Threshold' in line[0]:
            MAPQ.append(int(line[1].strip().replace(",", "").split('(')[0].strip()))
        elif 'Hi-C Contacts' in line[0]:
            hic.append(int(line[1].strip().replace(",", "").split('(')[0].strip()))
        elif 'Inter-chromosomal' in line[0]:
            inter.append(int(line[1].strip().replace(",", "").split('(')[0].strip()))
        elif 'Intra-chromosomal' in line[0]:
            intra.append(int(line[1].strip().replace(",", "").split('(')[0].strip()))
        elif 'Short Range (<20Kb)' in line[0]:
            short.append(int(line[1].strip().replace(",", "").split('(')[0].strip()))
        elif 'Long Range (>20Kb)' in line[0]:
            long.append(int(line[1].strip().replace(",", "").split('(')[0].strip()))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Your data
threshold = 0.6

# Create an array of indices for the bars
indices = ['Cortex_1','Cortex_2',
          'Medulla_1','Medulla_2',
          'Papilla_1','Papilla_2']
color_bar=['#E29412','#E29412','#1F7BE7','#1F7BE7','#D0021B','#D0021B']

# Set the figure size
fig, ax = plt.subplots(figsize=(5,3),dpi=300)

# Plot the bar chart for values
ax.bar(indices, hic_fraction, color=color_bar)

# Add a threshold line
ax.axhline(y=threshold, color='r', linestyle='--', label='High quality threshold (MIN)')

# Customize the plot
#ax.set_title('Bar Chart of Values with Threshold Line')
#ax.set_xlabel('Index')
ax.set_ylabel('#Hi-C contacts /\n#Alignable reads')
ax.set_ylim(0,0.9)
ax.set_xticklabels(rotation=45, labels=indices)
ax.legend(loc='lower right', bbox_to_anchor=(1, 0), frameon=True)

# Show the plot

plt.show()
```
